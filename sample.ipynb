# Generate an image with url

```python
client = OpenAI()

model_name = "dall-e-3"
prompt = """
"""
size="1024x1024"
quality="standard"
n=1
response_format="url"

response = client.images.generate(
    model=model_name,
    prompt=prompt,
    size=size,
    quality=quality,
    response_format=response_format,
    n=n,
)

image_url = response.data[0].url
image_url
```

# Generate image by b64

```python
response = requests.get(image_url)
img_bytes = response.content
type(img_bytes)
```
